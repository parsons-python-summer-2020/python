## Pandas

for this live code exercise we are going to use the 311 dataset from [NYC Open Data](https://opendata.cityofnewyork.us/). we'll be using the Pandas library to prepare the data that we'll plot later using Matplotlib.


```python
# importing both numpy and pandas
import numpy as np
import pandas as pd

# we are adding this line of code so that the columns don't collapse when it reaches a certain #
# this way we can see all of the columns all at once
pd.set_option('display.max_columns', None)
```

```python
# reading the csv file into a pandas dataframe
df = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv')
```

```python
# let's inspect our dataframe
# see the first 10 rows
# .tail(x) will show you the bottom x rows
df.head(10)
```

```python
# print a list of the column headers
# there are other ways of doing this but this is the simplest one!
list(df)
```

```python
# some of these columns don't have any real values that we can work with, so we'll drop them
# from our dataframe:
# axis = 1 refers to column, inplace=True will replace the current dataframe with the updated one
df.drop(['Address Type', 'Facility Type', 'Due Date', 'BBL', 'Vehicle Type', 
         'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 
         'Bridge Highway Direction', 'Road Ramp', 'Bridge Highway Segment',], axis=1, inplace=True)
```

```python
# show the first 10 rows
df.head(10)
```

```python
# let's change the zipcode from floats to integers
# replace numpy NaNs (Not a Number) with 0s
# change floats to integers
df['Incident Zip'] = df['Incident Zip'].replace(np.nan, 0)
df['Incident Zip'] = df['Incident Zip'].astype(int)
```

```python
# show the first 10 rows
df.head(10)
```

```python
# store new dataframe of 311 calls from Brooklyn
brooklyn = df[df['Borough'] == 'BROOKLYN']
brooklyn
```

```python
# size of brooklyn (row, column)
brooklyn.shape
```

```python
# reset the index
brooklyn.reset_index(drop=True, inplace=True)
brooklyn
```

```python
# store new dataframe of 311 calls from Manhattan
manhattan = df[df['Borough'] == 'MANHATTAN']
manhattan.reset_index(drop=True, inplace=True)
manhattan
```

```python
# size of manhattan (row, column)
manhattan.shape
```

```python
# meeting multiple conditions
fireworks_queens = df[(df['Borough'] == 'QUEENS') & (df['Complaint Type'] == 'Illegal Fireworks')]
fireworks_queens.reset_index(drop=True, inplace=True)
fireworks_queens.shape
```

```python
# show the output
fireworks_queens
```

```python
# loc gets rows (or columns) with particular labels from the index. 
# iloc gets rows (or columns) at particular positions in the index (so it only takes integers)
fireworks_queens.iloc[[4]]
```

```python
# slicing by rows
fireworks_queens[5:9]
```

```python
# select particular cell
fireworks_queens.iloc[1, 5]
```

```python
# example of column slicing
# : --> all rows
key_date = fireworks_queens.iloc[:, 0:2]
key_date
```

```python
# creating a copy with specific columns
select_columns = fireworks_queens[['Borough', 'Complaint Type', 'Location Type']]
select_columns
```

```python
# let's get the data that we need for our datavis project
# we want the boroughs, complaint type, location type, latitude and the longitude
# we also just want 'Complaint Type' to include the word 'Noise'
# first we filter through our entire dataframe to find the data that meets our criteria
noise = df[df['Complaint Type'].str.contains('Noise')]
noise
```

```python
# now we can select the columns we need
# reset the index
# and we're done!
noise_filtered = noise[['Borough', 'Complaint Type', 'Location Type', 'Latitude', 'Longitude']]
noise_filtered.reset_index(drop=True, inplace=True)
noise_filtered
```
