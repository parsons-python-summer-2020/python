## Convert dialogue to token list

```python
dialogue_txt.split(' ')
```

```python
import re

def prepareText(text):
    # Replace all end punctuation with space
    # pattern = r"((^|[^a-z])')|('($|[^a-z]))|([^a-z'])"
    pattern = r"[^a-z]"
    prepared_text = re.sub(pattern, ' ', text.lower())
    
    # Return the prepared text
    return prepared_text

prepareText("Wow...would you look at that!!!")
```

```python
all_tokens = re.split(r" +", prepareText(dialogue_txt))
```

## Unique the list

```python
def deduplicate(token_list, sort=True):
    unique_tokens = []

    for token in token_list:
        if token not in unique_tokens:
            unique_tokens.append(token)
    
    # By defaut, return a sorted list
    # This can be overridden in the function call
    if (sort):
        unique_tokens.sort()
    
    return unique_tokens
```

```python
uniuqe_tokens = deduplicate(all_tokens)
```

## Write to file

```python
with open('seinfeld_tokens.txt', 'w') as f:
    for token in unique_tokens:
        f.write(token + '\n')
```
